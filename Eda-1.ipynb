```python
Fixed Acidity: 
    The concentration of non-volatile acids in the wine, which contributes to its overall acidity.
Volatile Acidity: 
    The concentration of volatile acids in the wine, which can contribute to unpleasant aromas and flavors if present in excess.
Citric Acid:
    The concentration of citric acid, which can add a fresh and citrusy flavor to the wine.
Residual Sugar:
    The amount of sugar left over after fermentation is complete, which affects the sweetness of the wine.
Chlorides: 
    The concentration of salt in the wine, which can impact its taste and mouthfeel.
Free Sulfur Dioxide: 
    The amount of sulfur dioxide in a free form, which acts as an antioxidant and preservative.
Total Sulfur Dioxide:
    The total amount of sulfur dioxide, including both free and bound forms. Sulfur dioxide is important for wine preservation.
Density:
    The density of the wine, which can provide information about its composition.
pH: 
    The measure of acidity or alkalinity of the wine, which affects its stability and taste.
Sulphates:
    The concentration of sulphates, which can contribute to the wine's overall flavor and aroma.
Alcohol: 
    The alcohol content of the wine, which impacts its body, flavor, and overall character.
Quality:
    The sensory score assigned by experts or tasters to indicate the overall quality of the wine.
```

```python
import scipy.stats as stat
def my_QQplot(data,feature):
    
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    plt.hist(data[feature], bins=20)
    plt.subplot(1,2,2)
    stat.probplot(data[feature] , dist="norm" , plot=plt)
    plt.xlabel(feature)
    plt.show()
```

## Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number offeatures. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
df =  pd.read_csv(r"C:\Users\dhruv\Downloads\winequality-red.csv",sep=";")
df
```

```python
x = df.drop("quality" ,axis=1)
```

```python
y = df[["quality"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler_data = scaler.fit_transform(x_train)
```

```python
scaler_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5)
```

```python
pca_data = pca.fit_transform(scaler_data)
```

```python
pca_data
```

```python
ratio = pca.explained_variance_ratio_
```

```python
ratio.cumsum()[len(ratio)-1]
```

```python
pca = PCA(n_components=6)
pca_data = pca.fit_transform(scaler_data)
ratio = pca.explained_variance_ratio_
```

```python
ratio.cumsum()[len(ratio)-1]
```

```python
pca = PCA(n_components=7)
pca_data = pca.fit_transform(scaler_data)
ratio = pca.explained_variance_ratio_
```

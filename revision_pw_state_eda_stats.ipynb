```python
data = sns.load_dataset("tips")
```

```python
data 
```

```python
data[["sex","smoker"]]
```

```python
## compare two categorical we use chi square test
```

```python
data2 = pd.crosstab(data["sex"],data["smoker"])
data2
```

```python
observed_value = data2.values
observed_value
```

```python
import scipy.stats as stats
```

```python
stats_test,p,dof,expected_value = stats.chi2_contingency(observed_value)
```

```python
p
```

```python
dof
```

```python
stats_test
```

```python
expected_value
```

```python
data2
```

```python
no_row = data2.iloc[0:2,0]
len(no_row)
```

```python
no_columns = data2.iloc[0,0:2]
len(no_columns)
```

```python
dof = (len(no_row)-1)*(len(no_columns)-1)
dof
```

```python
observed_value
```

```python
expected_value
```

```python
zip(observed_value,expected_value)
```

```python
for o,e in zip(observed_value,expected_value):
    print(o)
    print(e)
```

```python
chi_square_test = sum([(o-e)**2/e for o,e in zip(observed_value,expected_value)])
```

```python
chi_square_test
```

```python
chi_statistics = chi_square_test[0] + chi_square_test[1]
```

```python
chi_statistics
```

```python
## chi_critical
```

```python
print(dof)
alpha = 0.05
```

```python
critical_value = stats.chi2.ppf(1-alpha,df=dof)
```

```python
critical_value
```

```python
p_val = stats.chi2.cdf(x=chi_statistics,df=dof)
p_val
```

```python
if chi_statistics>=critical_value:
    print("reject null hypothesis")
else :
    print("fail to reject null hypothesis")
    
```

```python
if p_val <= alpha:
    print("we will reject the null hypothesis")
else :
    print("we accept the null hypothesis")
```

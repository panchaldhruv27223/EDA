```python
def duplicated(df):
    duplicated = {}
    seen_columns = {}
    
    for columns in df.columns :
        current_columns = df[columns]
        
        try:
            current_columns_hash = current_columns.values.tobytes()
        except AttributeError:
            current_columns_hash = current_columns.to_string().encode()
            
        if current_columns_hash in seen_columns:
            if seen_columns[current_columns_hash] in duplicated:
                duplicated[seen_columns[current_columns_hash]].append(columns)
            else :
                duplicated[seen_columns[current_columns_hash]] = [columns]
        else :
            seen_columns[current_columns_hash] = columns
            
    return duplicated
```
